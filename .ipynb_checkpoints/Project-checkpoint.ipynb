## Part 2: Correlation Analysis

In this task, I conduct a correlation analysis to understand the relationships between different features in the dataset and their influence on music genres

```python
# Remove the rows with missing genre labels
df_music_realistic_nonull = df_music_realistic.dropna()
df_music_realistic_nonull.head()
```

```python
df_music_realistic_nonull.info()
```

```python
# Split the data into features and target
X = df_music_realistic_nonull.drop(columns=['Genre'])
y = df_music_realistic_nonull['Genre']
```

```python
# Encode the target labels - transforms categorical labels into numerical labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
# Update the dataframe - Replace non-encoded 'Genre' column with encoded 'Genre' column
df_with_encoded_genre = df_music_realistic_nonull.copy()
df_with_encoded_genre.drop(columns=['Genre'], inplace=True)
df_with_encoded_genre['Genre_encoded'] = y_encoded
df_with_encoded_genre.head()
```

```python
# Calculate pairwise correlation matrix for the dataframe with numeric 'Genre' values
correlation_matrix = df_with_encoded_genre.corr()
correlation_matrix
```

```python
# Visualize the correlation matrix using sns.heatmap - heatmap is used for visualizing matrix-like data
plt.figure(figsize=(15,10))
sns.heatmap(correlation_matrix,
            vmin=-1,
            vmax=1,
            annot=True,
            cmap='coolwarm',
            linewidths=0.1)
plt.title('Correlation matrix')
plt.show()
```

## Part 3: PCA for Dimensionality Reduction

In this task, I apply PCA to the dataset for dimensionality reduction. My objective is to determine the number of principal components necessary to capture at least 80% of the data variance by examining the cumulative explained variance plot.

```python
# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply PCA without specifying the number of components initially. This way, I'll obtain the same number of components as the number of features
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
# Analyze variance
explained_variance = pca.explained_variance_ratio_

# Calculate cummulative explained variance
cummulative_explained_variance = explained_variance.cumsum()
```

```python
# Visualize Cummulative Variance using Scree plot
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance) + 1), cummulative_explained_variance, marker='o', linestyle='-')
plt.xlabel('Number of components')
plt.ylabel('Cummulative explained variance')
plt.show()
```

My objective was to identify the minimum number of components necessary to capture at least 80% of the total variance. Therefore, according to the plot, I can use the first eight principal components.

```python
# After deciding the number of components, perform PCA again by defining n_components = 8, and transform the scaled data with this PCA model
pca = PCA(n_components=8)
X_pca = pca.fit_transform(X_scaled)
```

## Part 4: Evaulating Classification Efficacy: PCA-Transformed Vs. Original Data

In this task, I train and evaluate a Logistic Regression classifier using the principal components obtained from PCA. I compare its performance with a classifier trained on the original, non-PCA-transformed features.

### A. Logistic Regresssion with PCA-Transformed Data

```python
# Split the PCA-Transformed data and the encoded target variable into training and test sets into 70/30
X_train, X_test, y_train, y_test = train_test_split(X_pca, y_encoded, test_size=0.3, random_state=42)
```

```python
# Train a LogisticRegression classifier with PCA transformed data
classifier = LogisticRegression(max_iter=10000)
classifier.fit(X_train, y_train)
```

```python
# Predict and Evaluate the model.
y_pred = classifier.predict(X_test)

# Assess the model's accuracy
print(f"Accuracy with PCA: {accuracy_score(y_test, y_pred)}")

print(f"\nClassification Report:\n {classification_report(y_test, y_pred, target_names=le.classes_)}")
```

### B. Logisitic Regression with Original Features

```python
# Split the original (non-PCA) data and the encoded target variable into training and test sets into 70/30
X_train_orig, X_test_orig, y_train_orig, y_test_orig = train_test_split(X_scaled, y_encoded, test_size=0.3, random_state=42)
```

```python
# Train a logisticRegression model with original features(non-PCA transformed data)
classifier_orig = LogisticRegression(max_iter=10000)
classifier_orig.fit(X_train_orig, y_train_orig)
```

```python
# Predict and Evaluate the model with original features
y_pred_orig = classifier_orig.predict(X_test_orig)

# Assess the model's accuracy
print(f"Accuracy with Original Features: {accuracy_score(y_test_orig, y_pred_orig)}")

print(f"\nClassification Report with Original Features:\n {classification_report(y_test_orig, y_pred_orig, target_names=le.classes_)}")
```

Based on the evaluations of above two logisitic regression models, here’s an analysis and contrast of the accuracy and classification report:

* Both models have similar overall accuracies but show variations in their performance on individual classes.

## Part 5: Genre Prediction and Integration

In this task, I leverage the more effective model (PCA or original features) to predict unknown genres in the dataset. I apply these predictions to fill in the missing genre information in the original DataFrame, completing the dataset.

```python
# Locate and separate the rows in original dataframe that lack 'Genre' information and store it in a new dataframe
df_unknown_genre = df_music_realistic[df_music_realistic['Genre'].isnull()].copy()
df_unknown_genre
```

```python
# Prepare the data for prediction (scale the data to standardize it and transform it using PCA to match the format used for model training phase)
X_unknown = df_unknown_genre.drop(columns=['Genre'])
X_unknown_scaled = scaler.transform(X_unknown)
X_unknown_pca = pca.transform(X_unknown_scaled)

# Predict the genres using the previously selected and trained classifier
y_unknown_pred = classifier.predict(X_unknown_pca)
```

```python
# Transform the numerical predicted genres back to categorical values and assign them back to the DataFrame using lableEncoder.inverse_transform()
df_unknown_genre.loc[:, 'Predicted_Genre'] = le.inverse_transform(y_unknown_pred)
```

```python
# Display the DataFrame with predicted genres to verify that the model's predictions have correctly inserted the previously missing genre data
print(df_unknown_genre[['Predicted_Genre']])
```
